#### TRAINING AND EXTRACTION

```python
def extract_names(obj):
    if isinstance(obj, list):
        return [d.get("name", "") for d in obj if isinstance(d, dict)]
    return []

df["genres"] = df["genres"].apply(extract_names)
df["keywords"] = df["keywords"].apply(extract_names)
df["spoken_languages"] = df["spoken_languages"].apply(extract_names)
```

```python
def create_soup(row):
    return (
        " ".join(row['genres']) + " " +
        " ".join(row['keywords']) + " " +
        " ".join(row['spoken_languages']) + " " +
        str(row['tagline']) + " " +
        str(row['overview'])
    )

df["soup"] = df.apply(create_soup, axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Convert text to matrix
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df['soup'])

# Compute similarity
cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
indices = pd.Series(df.index, index=df['title']).drop_duplicates()
```

```python
def recommend(title, cosine_sim=cosine_sim):
    idx = indices[title]  # movie index
    
    # Get similarity scores for this movie
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sort by similarity
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Get top 10 recommendations (excluding itself)
    sim_scores = sim_scores[1:11]
    
    # Get indices of similar movies
    movie_indices = [i[0] for i in sim_scores]
    
    return df['title'].iloc[movie_indices]
```

```python
# Interactive recommendation
while True:
    user_input = input("Enter a movie name (or type 'exit' to quit): ")
    
    if user_input.lower() == "exit":
        print("Goodbye!")
        break
    
    # Check if movie exists in your dataset
    if user_input not in df['title'].values:
        print("Movie not found. Please try another title.")
        continue
    
    # Get recommendations
    recommended_movies = recommend(user_input)
    
    print(f"\nSince you liked '{user_input}', you may also like:")
    for i, movie in enumerate(recommended_movies, start=1):
        print(f"{i}. {movie}")
    print("\n" + "-"*40 + "\n")
```
